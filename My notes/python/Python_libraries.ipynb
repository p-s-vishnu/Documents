# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.title('Test')
plt.xlabel("Indices")
plt.ylabel("Numbers")
plt.grid()

plt.plot([2, 3, 6, 8])
plt.show()
```

```python
# blue line, red sqaure, green triangle
```

```python
a = np.arange(.1, 1.8, .2)

plt.plot(a, a**2, 'b--', label='^2')
plt.plot(a, a**3, 'rs', label='^3')
plt.plot(a, a**4, 'g^', label='^4')
plt.grid()
plt.legend()
plt.show()
```

```python
# line object - plt.plot(x1, y1, x2, y2)

line = plt.plot([1,2,3],[4,6,8], [4,5,6], [8,10,12])
plt.setp(line[1], color='r', linewidth=3.0)
plt.grid()
plt.show()
```

```python
# subplots
```

```python
def plot_fig(x):
    return np.sin(x) + np.cos(x)

plt.figure(1)
plt.subplot(121)
x = np.arange(10)
plt.plot(plot_fig(x), 'r')
# plt.grid()
# plt.show()

plt.subplot(122)
plt.plot(plot_fig(-x))
plt.grid()
plt.show()
```
